# **Лабораторная работа №4: Выделение контуров на изображении**
## **Цель работы**
Цель данной лабораторной работы – изучить и реализовать методы выделения контуров на изображении с использованием градиентных операторов.  

**В рамках работы выполняются следующие задачи:**  
- Преобразование исходного цветного изображения в полутоновое.  
- Вычисление градиентов по направлениям X и Y с помощью оператора Прюитта.  
- Нормализация полученных градиентных изображений.  
- Вычисление итогового градиента.  
- Применение бинаризации для выделения контуров.  

```python
import matplotlib.pyplot as plt
import cv2

# Загружаем изображения (цветное, полутоновое, Gx)
color_image = cv2.imread("0_image.png")
color_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB)

gray_image = cv2.imread("0_gray_image.png", cv2.IMREAD_GRAYSCALE)
gradient_x = cv2.imread("0_iGradient_X.png", cv2.IMREAD_GRAYSCALE)

# Отображаем первый набор изображений
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(color_image)
axes[0].set_title("Оригинальное изображение")
axes[0].axis("off")

axes[1].imshow(gray_image, cmap="gray")
axes[1].set_title("Полутоновое изображение")
axes[1].axis("off")

axes[2].imshow(gradient_x, cmap="gray")
axes[2].set_title("Градиент по X (Gx)")
axes[2].axis("off")

plt.show()

# Загружаем изображения (Gy, итоговый градиент, бинарное изображение)
gradient_y = cv2.imread("0_iGradient_Y.png", cv2.IMREAD_GRAYSCALE)
gradient_result = cv2.imread("0_iGradient_Result.png", cv2.IMREAD_GRAYSCALE)
gradient_binary = cv2.imread("0_iGradient_Binary.png", cv2.IMREAD_GRAYSCALE)

# Отображаем второй набор изображений
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(gradient_y, cmap="gray")
axes[0].set_title("Градиент по Y (Gy)")
axes[0].axis("off")

axes[1].imshow(gradient_result, cmap="gray")
axes[1].set_title("Результирующий градиент (G)")
axes[1].axis("off")

axes[2].imshow(gradient_binary, cmap="gray")
axes[2].set_title("Бинаризованное изображение")
axes[2].axis("off")

plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2

# Загружаем изображения (полутоновое, Gx, Gy)
gray_image = cv2.imread("gray_image.png", cv2.IMREAD_GRAYSCALE)
gradient_x = cv2.imread("gray_Gradient_X.png", cv2.IMREAD_GRAYSCALE)
gradient_y = cv2.imread("gray_Gradient_Y.png", cv2.IMREAD_GRAYSCALE)

# Создаем первый набор графиков
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(gray_image, cmap="gray")
axes[0].set_title("Полутоновое изображение")
axes[0].axis("off")

axes[1].imshow(gradient_x, cmap="gray")
axes[1].set_title("Градиент по X (Gx)")
axes[1].axis("off")

axes[2].imshow(gradient_y, cmap="gray")
axes[2].set_title("Градиент по Y (Gy)")
axes[2].axis("off")

plt.show()

# Загружаем следующие изображения (G, бинаризованное изображение)
gradient_result = cv2.imread("gray_Gradient_Result.png", cv2.IMREAD_GRAYSCALE)
gradient_binary = cv2.imread("gray_Gradient_Binary.png", cv2.IMREAD_GRAYSCALE)

# Создаем второй набор графиков
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(gradient_result, cmap="gray")
axes[0].set_title("Результирующий градиент (G)")
axes[0].axis("off")

axes[1].imshow(gradient_binary, cmap="gray")
axes[1].set_title("Бинаризованное изображение")
axes[1].axis("off")

plt.show()
```
