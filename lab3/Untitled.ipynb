## Фильтрация изображений и морфологические операции  

В данной лабораторной работе выполнялось морфологическое закрытие изображений с использованием структурированного элемента в форме кольца.  

Были получены следующие изображения:  
- **Оригинальное изображение**  
- **Обработанное изображение (морфологическое закрытие)**  
- **Разностное изображение (модуль разницы между оригиналом и закрытым изображением)**


```python
import matplotlib.pyplot as plt
import cv2

original = cv2.imread("1_79.png", cv2.IMREAD_GRAYSCALE)
closed = cv2.imread("1closed_79.png", cv2.IMREAD_GRAYSCALE)
difference = cv2.imread("1difference_79.png", cv2.IMREAD_GRAYSCALE)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(original, cmap="gray")
axes[0].set_title("Оригинальное изображение")
axes[0].axis("off")

axes[1].imshow(closed, cmap="gray")
axes[1].set_title("Морфологическое закрытие")
axes[1].axis("off")

axes[2].imshow(difference, cmap="gray")
axes[2].set_title("Разностное изображение")
axes[2].axis("off")

plt.show()
```
